```python
# Question 3
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# Data set contains independent variable and dependent variable for boston. The data is continuous type
## The dependent variable is in continuous type so we have to apply Linear regression for the prediction

```python
data=load_boston()
```

```python
# Question 4
x=data['data']
y=data['target']
print(x.shape)
```

```python
print(y.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
f_com=pd.DataFrame({"Actual":y_test,"M_predicted":y_pred})
```

```python
f_com
```
